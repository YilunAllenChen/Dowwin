# Stable Baselines DQN example

```python
import gym

from stable_baselines import DQN
from stable_baselines.common.evaluation import evaluate_policy
```

```python
# grab an environment
env = gym.make('CartPole-v1')
# grab a model
model = DQN('MlpPolicy', env, learning_rate=1e-3, prioritized_replay=True,
            verbose=1)
model.learn(total_timesteps=1000)
```

```python
# view the training result

observations = env.reset()

for i in range(1000):
    action, states = model.predict(observations)
    observations, rewards, dones, info = env.step(action)
    env.render()
    if dones:
        observations = env.reset()
```
